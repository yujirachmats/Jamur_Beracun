```python
df = pd.read_csv('mushrooms.csv')
```

```python
df
```

```python
# check null values
df.isnull().sum()
```

### Mengubah data menjadi Categorical data

- Menggunakan LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
for x in df.columns:
    df[x] = label.fit_transform(df[x])

df.head()
```

```python
# Melihat korelasi features x dengan target y (class)
corr = df.corr()
```

```python
# melihat penyebaran distribusi korelasi pada heatmap
plt.figure(figsize=(20,10))
k = 23 #number of variables for heatmap
cols = corr.nlargest(k, 'class')['class'].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

### Train Test Split 80% - 20%

```python
from sklearn.model_selection import train_test_split

xtr, xts, ytr, yts = train_test_split(df[df.columns[1:24]], df['class'], test_size=0.2)
print(len(xtr),len(xts),len(ytr),len(yts))
```

### Menggunakan Model Binary Classification dengan features CategoricaL Data

- Dengan menggunakan RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# default
model = RandomForestClassifier()
model.fit(xtr, ytr)
```

```python
# prediksi
model.predict(xts)
```

```python
df['prediksi'] = model.predict(df[df.columns[1:24]])
```

```python
# mengubah order column agar prediksi berada disebelah class
cols = df.columns.tolist()
# cols = cols[-1:] + cols[:-1]
# df = df[cols]
df
```

```python
from sklearn.metrics import (
    accuracy_score, recall_score, precision_score, 
    f1_score, balanced_accuracy_score, precision_recall_fscore_support)
```

1. __Accuracy Score__ & __Error / Misclassification Rate__

```python
print('akurasi = ', accuracy_score(df['class'], df['prediksi']))
print('Error Rate', round(1 - accuracy_score(df['class'], df['prediksi'])))
```

2. __Recall(+) / TP Rate & Recall(-) / TN Rate__

```python
print('recall(+) = ', recall_score(df['class'], df['prediksi'], pos_label=1))
print('recall(-) = ', recall_score(df['class'], df['prediksi'], pos_label=0))
```

3. __Precision(+) & Precision(-)__

```python
print('Precision(+)', precision_score(df['class'], df['prediksi'],pos_label=1))
print('Precision(-)', precision_score(df['class'], df['prediksi'],pos_label=0))
```

4. __Balanced Accuracy (bACC)__

```python
print('bACC = ',balanced_accuracy_score(df['class'], df['prediksi']))
```

5. __F1 Score__

```python
print('f1_score(+) = ', f1_score(df['class'], df['prediksi'], pos_label=1))
print('f1_score(-) = ', f1_score(df['class'], df['prediksi'], pos_label=0))
```

### ROC AUC

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
# plot ROC curve
# roc_curve(df['y'], df['proba'])
fpr, tpr, thres = roc_curve(yts, model.predict_proba(xts)[:,1])

plt.plot(fpr, tpr)
plt.plot([0,1], [0,1], 'r:')
plt.xlabel('False Positive rate')
plt.ylabel('True Positive rate')
plt.show()
```

```python
# luas AUC
roc_auc_score(yts, model.predict(xts))
```
